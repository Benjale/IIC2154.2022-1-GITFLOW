```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from pandas.io.json import json_normalize
dataFrame = pd.read_json('farmers-protest-tweets-2021-03-5.json', lines=True)
```

```python
users = json_normalize(dataFrame['user'])
users.drop(['description', 'linkTcourl'], axis=1, inplace=True)
users.rename(columns={'id':'userId', 'url':'profileUrl'}, inplace=True)
users = pd.DataFrame(users)
users.drop_duplicates(subset=['userId'], inplace=True)
```

```python
# Transform 'raw_tweets' DataFrame

# Add column for 'userId'
user_id = []
for user in dataFrame['user']:
    uid = user['id']
    user_id.append(uid)
dataFrame['userId'] = user_id

# Remove less important columns
cols = ['url', 'date', 'renderedContent', 'id', 'userId', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount', 'source', 'media', 'retweetedTweet', 'quotedTweet', 'mentionedUsers']
tweets = dataFrame[cols]
tweets.rename(columns={'id':'tweetId', 'url':'tweetUrl'}, inplace=True)
tweets = pd.DataFrame(tweets)
tweets.drop_duplicates(subset=['tweetId'], inplace=True)
```

```python
import retweeted
most_retweeted = retweeted.top(tweets)
most_retweeted.head(10)
```

```python
import us
dfNew=pd.DataFrame(columns = ['username' , 'displayname', 'userId'])
top_users = us.top(tweets, dfNew, users)
top_users.head(10)
```

```python
import days
top_days = days.top(tweets)
top_days.head(10)
```
